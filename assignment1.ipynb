```python
import pandas as pd
```

```python
data=pd.read_csv('E:\python_numpy\sample_iowa_liquor_sales.csv')
data
```

```python
Sales Analysis by Store:

What is the total sales amount (Sale (Dollars)) per store?

How many bottles were sold (Bottles Sold) per store?

Which store has the highest total sales?

Which store sells the most volume in liters (Volume Sold (Liters))?
```

```python
Sales per store and    converting data to .excel
```

```python
new = data.groupby ('Store Name').agg(
    total_sale_value = ('Sale (Dollars)', 'sum'),
    total_bottle_sold = ('Bottles Sold' , 'sum'),
     storewise_volume_in_ltrs=('Volume Sold (Liters)','sum')
)
new
```

```python
Max highest sale:
```

```python
max_sale = new['total_sale_value'].max()
new.loc[new['total_sale_value']==max_sale]

```

```python
                                        SORT   A VALUE :-----  (ASCENDING     and     DESCENDING)
```

```python
                                        ASCENDING SORTING
```

```python
new.sort_values('total_sale_value')
```

```python
                                                DESCENDING    SORTING 
```

```python
new.sort_values('total_bottle_sold' , ascending=False)
```
